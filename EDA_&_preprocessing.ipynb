Feature engineering: When your trying to change whole column data type ,or your trying to change whole column values and that step is feature engineering 

Encoding : Label encoding :suitable for column values which is having a relation or an order 

```python
data['Price']=LabelEncoder().fit_transform(data['Price'])
```

```python
data['Open']=LabelEncoder().fit_transform(data['Open'])
```

```python
data['High']=LabelEncoder().fit_transform(data['High'])
```

```python
data['Low']=LabelEncoder().fit_transform(data['Low'])
```

```python
data['Vol.']=LabelEncoder().fit_transform(data['Vol.'])
```

```python
plt.figure(figsize=(10,5))
plt.pie(data["year"].value_counts(),labels=data["year"].unique(),autopct="%1.1f%%")
plt.show()
# Pie chart : Only with matplotlib
```

```python
plt.figure(figsize=(12,8))
sns.kdeplot(data[["Price","Vol."]])
plt.show()
# KDE(Kernal density estimation) PLOT:To understand data is normally distributed(Bell curve) or not 
```

```python
plt.figure(figsize=(12,8))
sns.histplot(data[["year","Price"]])
plt.show()
# Histplot : helps in data distribution and counting (works better with categorical columns ) 
```

```python
plt.figure(figsize=(10, 4))
plt.plot(data["Date"], data["Price"])
plt.title("Reliance Price Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.hist(data["Price"], bins=30, edgecolor="black")
plt.title("Reliance Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
# Histplot :  helps in data distribution and counting (works better with categorical columns ) 
```

```python
plt.figure(figsize=(10, 4))
plt.plot(data["Date"], data["Open"], label="Open")
plt.plot(data["Date"], data["High"], label="High")
plt.plot(data["Date"], data["Low"],  label="Low")
plt.title("Open, High, Low Prices Over Time")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.tight_layout()
plt.show()
```
